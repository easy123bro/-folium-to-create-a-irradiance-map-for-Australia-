# creat folium map  (deep colour reprents large value while light colour reprents    samll value

```python
import folium
import numpy as np

# 读取 CSV 文件的指定列
columns_to_read = ['latitude', 'longitude', 'CLOT']

# 从 CSV 加载数据
df = pd.read_csv('australian_wave_data.csv', usecols=columns_to_read)

# 提取纬度、经度和 CLOT 值
latitude_values = df['latitude'].values
longitude_values = df['longitude'].values
CLOT_values = df['CLOT'].values

# 将 CLOT 值限制在 0 到 10 的范围内
CLOT_values_clipped = np.clip(CLOT_values, 0, 1000)

# 创建 Folium 地图
m = folium.Map(location=[np.mean(latitude_values), np.mean(longitude_values)], zoom_start=8)

# 自定义颜色映射函数
def get_color(value):
    if value < 700:
        return '#ffffcc'  # 浅蓝色，表示低值
    elif value < 850:
        return '#fdae6b'  # 橙色，表示中等值
    elif value < 1000:
        return '#e6550d'  # 深橙色，表示较高值
    else:
        return '#a50f15'  # 暗红色，表示非常高的值


# 向地图添加矩形
for lat, lon, value in zip(latitude_values, longitude_values, CLOT_values_clipped):
    square_side_length_km = 10.0
    lat_degree_length_km = 110.574
    square_side_length_deg = square_side_length_km / lat_degree_length_km
    bounds = [
        [lat - square_side_length_deg / 2, lon - square_side_length_deg / 2],
        [lat + square_side_length_deg / 2, lon + square_side_length_deg / 2]
    ]

    folium.Rectangle(
        bounds=bounds,
        color=None,
        fill=True,
        fill_opacity=0.7,
        fill_color=get_color(value),
        stroke=False
    ).add_to(m)

# 保存地图
m.save('promax.html')
print("Folium map has been created!")
```
